# Import necessary libraries

```python
import numpy as np
from skimage import io, color, filters
```

# Load test images

```python
img1 = io.imread('IMG.jpg')
img2 = io.imread('img2.jpg')
img3 = io.imread('img3.jpg')
img4 = io.imread('img4.jpg')
```

# Convert images to grayscale

```python
gray1 = color.rgb2gray(img1)
gray2 = color.rgb2gray(img2)
gray3 = color.rgb2gray(img3)
gray4 = color.rgb2gray(img4)
```

# Apply median filter to remove noise

```python
median1 = filters.median(gray1)
median2 = filters.median(gray2)
median3 = filters.median(gray3)
median4 = filters.median(gray4)
```

# Apply threshold to binarize image

```python
threshold1 = filters.threshold_otsu(median1)
threshold2 = filters.threshold_otsu(median2)
threshold3 = filters.threshold_otsu(median3)
threshold4 = filters.threshold_otsu(median4)
```

```python
binary1 = median1 > threshold1
binary2 = median2 > threshold2
binary3 = median3 > threshold3
binary4 = median4 > threshold4
```

# Compute the fraction of pixels that are oily

```python
oily_fraction1 = np.sum(binary1) / binary1.size
oily_fraction2 = np.sum(binary2) / binary2.size
oily_fraction3 = np.sum(binary3) / binary3.size
oily_fraction4 = np.sum(binary4) / binary4.size
```

# Print the results

```python
print('Image 1 oily fraction:', oily_fraction1)
print('Image 2 oily fraction:', oily_fraction2)
print('Image 3 oily fraction:', oily_fraction3)
print('Image 4 oily fraction:', oily_fraction4)
```

# Plotting binary images

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=2)
ax = axes.ravel()
ax[0].imshow(binary1, cmap=plt.cm.gray)
ax[0].set_title("Image 1")

ax[1].imshow(binary2, cmap=plt.cm.gray)
ax[1].set_title("Image 2")

ax[2].imshow(binary3, cmap=plt.cm.gray)
ax[2].set_title("Image 3")

ax[3].imshow(binary4, cmap=plt.cm.gray)
ax[3].set_title("Image 4")

for a in ax:
    a.axis('off')
    
plt.tight_layout()
plt.show()
```
